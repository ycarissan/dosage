```python
import random
import numpy as np

Ve=12e-3
U_Ve=0.3e-3
CB=0.1
U_CB=0.005
VA=20e-3
U_VA=0.3e-3

CA=[]
for i in range(100000) :
    Veexp=Ve+U_Ve*(1-random.random())
    CBexp=CB+U_CB*(1-random.random())
    VAexp=VA+U_VA*(1-random.random())
    CA.append(CBexp*Veexp/VAexp)

hist,bins=np.histogram(CA,bins=np.linspace(min(CA),max(CA),200))
```

```python
from matplotlib import pyplot
import matplotlib.pyplot as plt

_=plt.hist(CA, bins)
```
